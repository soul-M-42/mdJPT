```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

def load_mean_attrs(path):
    """ attr  [N, F]"""
    attrs = np.load(path)
    attrs = attrs.reshape(attrs.shape[0], -1, attrs.shape[-1])
    return np.mean(attrs, axis=1)

def compute_cross_corr(A, B):
    """ B  A  Pearson  [B, A]"""
    assert A.shape[1] == B.shape[1], "Time dimension must match"
    num_a, num_b = A.shape[0], B.shape[0]
    corr = np.zeros((num_b, num_a))
    for i in range(num_b):
        for j in range(num_a):
            corr[i, j], _ = pearsonr(B[i], A[j])
    return corr

def reorder_matrix(matrix, row_order=None, col_order=None):
    """"""
    if row_order is not None:
        matrix = matrix[row_order, :]
    if col_order is not None:
        matrix = matrix[:, col_order]
    return matrix

import matplotlib.pyplot as plt


import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def plot_heatmap(data, xlabel=None, ylabel=None, xtick_fontsize=35, ytick_fontsize=35, show_colorbar=True):
    cell_width = 2
    cell_height = 2

    n_rows, n_cols = data.shape
    figsize = (n_cols * cell_width, n_rows * cell_height)
    abs_max = np.abs(data).max()

    fig, ax = plt.subplots(figsize=figsize)
    cax = ax.imshow(data, aspect='auto', cmap='coolwarm', vmin=-abs_max, vmax=abs_max)

    if show_colorbar:
        #  colorbar 
        #  colorbar  1.5  0.3 
        cbar_ax = inset_axes(ax,
                             width=0.3,    # 
                             height=4,   # 
                             loc='right',      # 
                             borderpad=-3       # 
                            )
        cbar = fig.colorbar(cax, cax=cbar_ax, orientation='vertical')
        cbar.ax.tick_params(labelsize=30)

    if xlabel is not None:
        assert len(xlabel) == data.shape[1]
        ax.set_xticks(range(data.shape[1]))
        ax.set_xticklabels(xlabel, rotation=90, fontsize=xtick_fontsize)
    else:
        ax.set_xticks([])

    if ylabel is not None:
        assert len(ylabel) == data.shape[0]
        ax.set_yticks(range(data.shape[0]))
        ax.set_yticklabels(ylabel, fontsize=ytick_fontsize)
    else:
        ax.set_yticks([])

    ax.set_aspect('equal')

    fig.subplots_adjust(
        left=0.2 if ylabel is not None else 0.05,
        right=0.85,  #  colorbar 
        top=0.95,
        bottom=0.3 if xlabel is not None else 0.05
    )

    plt.show()



```

```python
# 
att_SEEDV = load_mean_attrs('importance/DATASET_loo_DEAP/SEEDV/attrs_allData_cls_all.npy')
att_SEED  = load_mean_attrs('importance/DATASET_loo_DEAP/SEED/attrs_allData_cls_all.npy')

# 
sim_matrix = compute_cross_corr(att_SEEDV, att_SEED)

# 
V_label = ['Disgust', 'Fear', 'Sad', 'Neutral', 'Happy']
SEED_label = ['Negative', 'Neutral', 'Positive']

# 
row_order = [2, 1, 0]
col_order = [4, 3, 2, 1, 0]

# 
sim_reordered = reorder_matrix(sim_matrix, row_order, col_order)

# 
SEED_label_new = [SEED_label[i] for i in row_order]
V_label_new = [V_label[i] for i in col_order]

# 
plot_heatmap(sim_reordered, xlabel=V_label_new, ylabel=None)

```
